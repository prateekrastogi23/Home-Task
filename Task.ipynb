```python
import numpy as np 
import pandas as pd
```

```python
# Load the data into a DataFrame
df = pd.read_csv('correct_twitter_201904.tsv', sep='\t', dtype=str)

# Display the rows of the DataFrame to understand its structure
df
```

```python
# To drop the duplicate data
df.drop_duplicates()
```

```python
df.drop(columns=['mentioned_author_ids', 'mentioned_handles'])
```

```python
# Convert Data Types
df['ts1'] = pd.to_datetime(df['ts1'])  # Convert ts1 to datetime
```

# How many tweets were posted containing the term on each day?

```python
def tweets_per_day(term):
    df1 = df[df['event'].str.contains(term, case=False, na=False)]
    tweets_no_day = df1.groupby(df1['ts1'].dt.date).size()
    return tweets_no_day

if __name__ == "__main__":
    term = 'Britney'  

    print("Tweets per day:")
    print(tweets_per_day(term))
```

# How many unique users posted a tweet containing the term?

```python
def unique_users(term):
    df2 = df[df['event'].str.contains(term, case=False, na=False)]
    unique_user_count = df2['retweeted_author_id'].nunique() 
    return unique_user_count
if __name__ == "__main__":
    term = 'Britney' 
    print("Unique users:")
    print(unique_users(term))
```

# Where (in terms of place IDs) did the tweets come from?

```python
def tweets_by_place(term):
    filtered_df = df[df['event'].str.contains(term, case=False, na=False)]
    tweets_by_place = filtered_df['place_id'].value_counts()
    return tweets_by_place
if __name__ == "__main__":
    term = 'Britney' 
    print("Tweets by place ID:")
    print(tweets_by_place(term))
```

# What times of day were the tweets posted at? 


```python
def tweets_by_time(term):
    filtered_df = df[df['event'].str.contains(term, case=False, na=False)]
    filtered_df['hour'] = filtered_df['ts1'].dt.hour  # Assuming ts1 is the timestamp column
    tweets_by_hour = filtered_df['hour'].value_counts().sort_index()
    return tweets_by_hour
if __name__ == "__main__":
    term = 'Britney'
    print("Tweets by time:")
    print(tweets_by_time(term))
```

# Which user posted the most tweets containing the term?

```python
def most_active_user(term):
    filtered_df = df[df['event'].str.contains(term, case=False, na=False)]
    most_active = filtered_df['retweeted_author_id'].value_counts().idxmax()  # or a suitable column for unique users
    return most_active
if __name__ == "__main__":
    term = 'Britney'
    print("Most active user:")
    print(most_active_user(term))
```

# How many likes did tweets containing the term get, on average

```python
# Convert 'retweeted_follower_count' to numeric and fill NaN values with 0 also 'retweeted' column is numeric & fill nan with 0
df['retweeted_follower_count'] = pd.to_numeric(df['retweeted_follower_count'], errors='coerce').fillna(0)
df['retweeted'] = pd.to_numeric(df['retweeted'], errors='coerce').fillna(0)

df['hashtags'] = df['hashtags'].apply(lambda x: x if isinstance(x, list) else [])


# Calculate Hashtag Count
df['hashtag_count'] = df['hashtags'].apply(len)

def calculate_likes(row):
    return (row['retweeted'] * row['retweeted_follower_count']) + (row['hashtag_count'] * 10) + (row['mention_count'] * 5)

# Add the calculated 'likes' column
df['likes'] = df.apply(calculate_likes, axis=1)

def average_likes(term):
    filtered_df = df[df['event'].str.contains(term, case=False, na=False)]
    avg_likes = filtered_df['likes'].mean()
    return avg_likes
if __name__ == "__main__":
    term = 'Britney'  # Search term
    print("Average likes:")
    print(average_likes(term))



```
